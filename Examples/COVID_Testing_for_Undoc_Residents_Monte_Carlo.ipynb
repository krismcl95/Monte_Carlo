# Code!

You can adjust the number of trials or the decimal level in the first cell. By default, you will run 10,000 trials and go to four decimal places.

If those defaults work for you, just run both cells!

```python
trials = 10000
dec_lev = 10000
```

```python
# create main dataframe
data_vals = []
data_cols = [A_name, B_name, C_name, D_name, E_name, F_name, G_name, H_name, I_name, J_name, K_name, L_name, u_name, v_name, w_name, x_name, y_name, z_name]
for i in range (0,trials):
        if A_dist == 'U':
            A = random.uniform(A_min, A_max)
        elif A_dist =='T':
            A = random.triangular(A_min, A_max, A_likely)
        if B_dist == 'U':
            B = random.uniform(B_min, B_max)
        elif B_dist =='T':
            B = random.triangular(B_min, B_max, B_likely)
        if C_dist == 'U':
            C = random.uniform(C_min, C_max)
        elif C_dist =='T':
            C = random.triangular(C_min, C_max, C_likely)
        if D_dist == 'U':
            D = random.uniform(D_min, D_max)
        elif D_dist =='T':
            D = random.triangular(D_min, D_max, D_likely)
        if E_dist == 'U':
            E = random.uniform(E_min, E_max)
        elif E_dist =='T':
            E = random.triangular(E_min, E_max, E_likely)
        if F_dist == 'U':
            F = random.uniform(F_min, F_max)
        elif F_dist =='T':
            F = random.triangular(F_min, F_max, F_likely)
        if G_dist == 'U':
            G = random.uniform(G_min, G_max)
        elif G_dist =='T':
            G = random.triangular(G_min, G_max, G_likely)
        if H_dist == 'U':
            H = random.uniform(H_min, H_max)
        elif H_dist =='T':
            H = random.triangular(H_min, H_max, H_likely)
        if I_dist == 'U':
            I = random.uniform(I_min, I_max)
        elif I_dist =='T':
            I = random.triangular(I_min, I_max, I_likely)
        if J_dist == 'U':
            J = random.uniform(J_min, J_max)
        elif J_dist =='T':
            J = random.triangular(J_min, J_max, J_likely)
        if K_dist == 'U':
            K = random.uniform(K_min, K_max)
        elif K_dist =='T':
            K = random.triangular(K_min, K_max, K_likely)
        if L_dist == 'U':
            L = random.uniform(L_min, L_max)
        elif L_dist =='T':
            L = random.triangular(L_min, L_max, L_likely)
        new_row = [A,B,C,D,E,F,G,H,I,J,K,L,int(eval(u)*dec_lev)/dec_lev,int(eval(v)*dec_lev)/dec_lev,int(eval(w)*dec_lev)/dec_lev,int(eval(x)*dec_lev)/dec_lev,int(eval(y)*dec_lev)/dec_lev,int(eval(z)*dec_lev)/dec_lev]
        data_vals.append(new_row)
df = pd.DataFrame(data=data_vals, columns=data_cols)

#create summary table
df_basic_info = df.describe()

#rename coulmns
df_basic_info.rename(columns={'A':A_name,
    'B': B_name, 
    'C': C_name, 
    'D': D_name,
    'E': E_name,
    'F': F_name,
    'G': G_name,
    'H': H_name,
    'I': I_name,
    'J': J_name,
    'K': K_name,
    'L': L_name,
    'u': u_name,
    'v': v_name,
    'w': w_name,
    'x': x_name,
    'y': y_name,
    'z': z_name}, inplace=True)

#add 5 and 95 percentiles
df_basic_info1 = df_basic_info[0:4] 
df_basic_info2 = df_basic_info[4:] 
df_basic_info1.loc['5%']=[df[A_name].quantile(0.05),df[B_name].quantile(0.05),df[C_name].quantile(0.05),df[D_name].quantile(0.05),df[E_name].quantile(0.05),df[F_name].quantile(0.05),df[G_name].quantile(0.05),df[H_name].quantile(0.05),df[I_name].quantile(0.05),df[J_name].quantile(0.05),df[K_name].quantile(0.05),df[L_name].quantile(0.05),df[u_name].quantile(0.05),df[v_name].quantile(0.05),df[w_name].quantile(0.05),df[x_name].quantile(0.05),df[y_name].quantile(0.05),df[z_name].quantile(0.05)]
df_basic_info = pd.concat([df_basic_info1, df_basic_info2])


df_basic_info1 = df_basic_info[0:8] 
df_basic_info2 = df_basic_info[8:] 
df_basic_info1.loc['95%']=[df[A_name].quantile(0.95),df[B_name].quantile(0.95),df[C_name].quantile(0.95),df[D_name].quantile(0.95),df[E_name].quantile(0.95),df[F_name].quantile(0.95),df[G_name].quantile(0.95),df[H_name].quantile(0.95),df[I_name].quantile(0.95),df[J_name].quantile(0.95),df[K_name].quantile(0.95),df[L_name].quantile(0.95),df[u_name].quantile(0.95),df[v_name].quantile(0.95),df[w_name].quantile(0.95),df[x_name].quantile(0.95),df[y_name].quantile(0.95),df[z_name].quantile(0.95)]
df_basic_info = pd.concat([df_basic_info1, df_basic_info2])

#return
df_basic_info
```

    Set the name of your dependent variable as result2_name

Run all five cells

    Note: The second one will take some time
    
Copy and paste the resulting table and graph to excel (or wherever)

    Potential Variable Reduction means that if you reduce the independent variable to a single point estimate, you can reduce the dependent variable's variance by approximately _%

```python
result2 = x
result2_name = 'Cost of Testing'
```

```python
data_cols2 = ['A_result2']
def sen_nums (i):
    global A
    global B
    global C
    global D
    global E
    global F
    global G
    global H
    global I
    global J
    global K
    global L
    A = df.loc[i, A_name]
    B = df.loc[i, B_name]
    C = df.loc[i, C_name]
    D = df.loc[i, D_name]
    E = df.loc[i, E_name]
    F = df.loc[i, F_name]
    G = df.loc[i, G_name]
    H = df.loc[i, H_name]
    I = df.loc[i, I_name]
    J = df.loc[i, J_name]
    K = df.loc[i, K_name]
    L = df.loc[i, L_name]

#A
data_vals2 = []
for i in range (0,trials):
    sen_nums(i)
    A = df[A_name].mean()
    new_row2 = [int(eval(result2)*dec_lev)/dec_lev]
    data_vals2.append(new_row2)
df_sen = pd.DataFrame(data=data_vals2, columns=data_cols2)

#B
data_vals3 = []
for i in range (0,trials):
    sen_nums(i)
    B = df[B_name].mean()
    data_vals3.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['B_result2'] = data_vals3

#C
data_vals4 = []
for i in range (0,trials):
    sen_nums(i)
    C = df[C_name].mean()
    data_vals4.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['C_result2'] = data_vals4

#D
data_vals5 = []
for i in range (0,trials):
    sen_nums(i)
    D = df[D_name].mean()
    data_vals5.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['D_result2'] = data_vals5

#E
data_vals6 = []
for i in range (0,trials):
    sen_nums(i)
    E = df[E_name].mean()
    data_vals6.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['E_result2'] = data_vals6

#F
data_vals7 = []
for i in range (0,trials):
    sen_nums(i)
    F = df[F_name].mean()
    data_vals7.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['F_result2'] = data_vals7

#G
data_vals8 = []
for i in range (0,trials):
    sen_nums(i)
    G = df[G_name].mean()
    data_vals8.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['G_result2'] = data_vals8

#H
data_vals9 = []
for i in range (0,trials):
    sen_nums(i)
    H = df[H_name].mean()
    data_vals9.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['H_result2'] = data_vals9

#I
data_vals10 = []
for i in range (0,trials):
    sen_nums(i)
    I = df[I_name].mean()
    data_vals10.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['I_result2'] = data_vals10

#J
data_vals11 = []
for i in range (0,trials):
    sen_nums(i)
    J = df[J_name].mean()
    data_vals11.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['J_result2'] = data_vals11

#K
data_vals12 = []
for i in range (0,trials):
    sen_nums(i)
    K = df[K_name].mean()
    data_vals12.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['K_result2'] = data_vals12

#L
data_vals13 = []
for i in range (0,trials):
    sen_nums(i)
    L = df[L_name].mean()
    data_vals13.append(int(eval(result2)*dec_lev)/dec_lev)
df_sen['L_result2'] = data_vals13

#Include original
df_sen['Original'] = df[result2_name]
```

```python
#Create Std Table
df_sen_std_vals = [
    1-(df_sen['A_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['B_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['C_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['D_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['E_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['F_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['G_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['H_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['I_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['J_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['K_result2'].std()/df_sen['Original'].std()),
    1-(df_sen['L_result2'].std()/df_sen['Original'].std())]
row_names = [A_name,B_name,C_name,D_name,E_name,F_name,G_name,H_name,I_name,J_name,K_name,L_name]
df_sen_std = pd.DataFrame({'Potential Variation Reduction': df_sen_std_vals},index= row_names).sort_values('Potential Variation Reduction', ascending = False)
df_sen_std = df_sen_std[df_sen_std['Potential Variation Reduction'] > 0]
```
